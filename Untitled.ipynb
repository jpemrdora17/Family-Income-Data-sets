<h1>10 Visualization </h1> <hr>

```python
df = pd.read_csv("Family Income and Expenditure.csv")
```

```python
df.head()
```

<h1>Number 1. What is the average total food expenditure per region?</h1>
<h2>Visualization: Bar Chart</h2>
<hr>

```python
avg_food = df.groupby('Region')['Total Food Expenditure'].mean().sort_values()


plt.figure(figsize=(15,6))
avg_food.plot(kind='bar', color='skyblue')
plt.title("Average Total Food Expenditure per Region")
plt.ylabel("Average Food Expenditure")
plt.xlabel("Region")
plt.xticks(rotation=45)

plt.show()
```

<h1>Insight: number 1 </h1>
<p>The chart shows that regions like NCR and CALABARZON have the highest food expenses, while regions like Northern Mindanao and ARMM have the lowest. This suggests that more urbanized areas tend to spend more on food, while rural areas spend less.
</p>

<h1>Number 2. How does total household income correlate with total food expenditure?</h1>
<h2>Visualization: Scatter Plot</h2>
<hr>

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Total Household Income'], df['Total Food Expenditure'], alpha=0.5)
plt.title("Household Income vs Total Food Expenditure")
plt.xlabel("Total Household Income")
plt.ylabel("Total Food Expenditure")
plt.show()
```

<h1>Insight: number 2 </h1>
<p>
The scatter plot shows that as household income increases, total food expenditure also tends to increase. Most households with lower income spend less on food, while households with higher income generally spend more. However, the points are spread out, which means not all families spend the same amount even if they have similar incomes. This suggests that income affects food spending, but other factors might also play a role.
</p>

<h1>Number 3. Which main source of income spends the most on rice?</h1>
<h2>Visualization: Box Plot</h2> <hr

```python
plt.figure(figsize=(10,6))
df.boxplot(column='Total Rice Expenditure', by='Main Source of Income', grid=False)
plt.title("Total Rice Expenditure by Main Source of Income")
plt.suptitle("")
plt.ylabel("Rice Expenditure")
plt.xticks(rotation=45)
plt.show()
```

<h1>Insight: number 3</h1>
<p>The chart shows that rice expenditure is quite similar across all three main sources of income: Entrepreneurial Activities, Other Sources of Income, and Wages/Salaries. Most households spend around the same amount on rice, no matter where their income comes from.</p>

<h1>Number 4. Heatmap of expenditure categories correlation </h1>

```python
# Select numeric columns related to expenditures
cols = ['Total Food Expenditure', 'Bread and Cereals Expenditure', 'Total Rice Expenditure', 
        'Meat Expenditure', 'Fruit Expenditure']

plt.figure(figsize=(10,8))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Expenditure Categories")

plt.show()
```

<h1>Insight: number 4</h1>
<h2>The heatmap illustrates the degree of relationship between various food expenditure categories.</h2>
<p>4.1. Total Food Expenditure is most strongly correlated with Meat Expenditure (0.77), indicating that meat spending is a significant factor in overall food spending.</p>
<p>4.2. The strongest internal correlation (0.88) exists between Bread and Cereals Expenditure and Total Rice Expenditure, suggesting that spending on these two staple categories is highly linked.</p>
<p>4.3. Fruit Expenditure shows the weakest correlation with other specific categories (like Rice at 0.12), implying that spending on fruits is the most independent of the other listed food types.</p>


<h1> number 5. Distribution of household income</h1>
<h2>Visualization: Histogram</h2>
<hr>

```python
plt.figure(figsize=(8,6))
plt.hist(df['Total Household Income'], bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title("Distribution of Total Household Income")
plt.xlabel("Income")
plt.ylabel("Number of Households")
plt.show()
```

<p>Household Income Distribution Insight
The visualization displays a highly right-skewed distribution of Total Household Income.
Vast Majority in Low-Income Bracket: The overwhelming majority of households (over 35,000) are concentrated in the lowest income bin (closest to zero).</p>
